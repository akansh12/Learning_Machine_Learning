```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
X = dataset.iloc[:,[3,4]].values
```

```python
import scipy.cluster.hierarchy as sch
```

```python
dendrogram = sch.dendrogram(sch.linkage(X,method = 'ward'))
```

```python
# we came to know that there shoul exist 5 clusters
from sklearn.cluster import AgglomerativeClustering
```

```python
hc = AgglomerativeClustering(n_clusters=5)
```

```python
y_hc= hc.fit_predict(X)
```

```python
y_hc
```

```python
plt.scatter(X[y_hc==0,0],X[y_hc==0,1],s=100,c='cyan')
plt.scatter(X[y_hc==1,0],X[y_hc==1,1],s=100,c='blue')
plt.scatter(X[y_hc==2,0],X[y_hc==2,1],s=100,c='red')
plt.scatter(X[y_hc==3,0],X[y_hc==3,1],s=100,c='black')
plt.scatter(X[y_hc==4,0],X[y_hc==4,1],s=100,c='magenta')
```
