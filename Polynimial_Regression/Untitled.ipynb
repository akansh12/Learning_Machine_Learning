```python
#Polynaomial Regression
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Position_Salaries.csv")
```

```python
X = dataset.iloc[:,1:2].values
y = dataset.iloc[:,2].values
```

```python
#Non Linear Graph
plt.scatter(dataset.Level,dataset.Salary)
```

```python
#Fitting linear regression
from sklearn.linear_model import LinearRegression
```

```python
lin_reg = LinearRegression()
```

```python
lin_reg.fit(X,y)
```

```python
y_pred = lin_reg.predict(X)
```

```python
plt.plot(X,y_pred,color = 'red')
plt.scatter(dataset.Level,dataset.Salary) 
```

```python
# Creating a Polynomial Model
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_reg = PolynomialFeatures(degree = 4)
```

```python
X_poly = poly_reg.fit_transform(X)
```

```python
X_grid = np.arange(0,10.01,0.1)
```

```python
X_grid = X_grid.reshape([len(X_grid),1])
```

```python
lin_reg_2 = LinearRegression()
```

```python
lin_reg_2.fit(X_poly,y)
```

```python
y_pred_2 = lin_reg_2.predict(X_poly)
```

```python
plt.plot(X,y_pred_2,color = 'green')
plt.plot(X,y_pred,color = 'red')
plt.scatter(dataset.Level,dataset.Salary) 
```

```python
plt.plot(X_grid,lin_reg_2.predict(poly_reg.fit_transform(X_grid)),color = 'green')
plt.plot(X,y_pred,color = 'red')
plt.scatter(dataset.Level,dataset.Salary) 
```
